# Default PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Iris Dataset")
plt.colorbar()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

# Randomized PCA

```python
rpca = PCA(n_components = 2, svd_solver = 'randomized', random_state = 42)
X_rpca = rpca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_rpca[:, 0], X_rpca[:, 1], c = y, cmap = 'tab10', alpha = 0.7)
plt.xlabel("Randomized Principal Component 1")
plt.ylabel("Randomized Principal Component 2")
plt.title("Digits dataset - Randomized PCA to 2D")
plt.colorbar(scatter, label='Digit Label')
plt.grid(True)
plt.show()
```
