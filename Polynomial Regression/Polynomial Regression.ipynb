```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = {
    'Mileage': [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000],
    'Price':   [500000, 480000, 455000, 420000, 380000, 340000, 310000, 285000, 260000, 240000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.scatter(df['Mileage'], df['Price'], color = 'green')
plt.xlabel('Mileage(KM)')
plt.ylabel('Price(Tk)')
plt.title('Mileage vs Price')
plt.show();
```

```python
X = df[['Mileage']]
y = df[['Price']]
```

```python
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
```

```python
sorted_zip = sorted(zip(X['Mileage'], y_pred))
X_sorted, y_sorted = zip(*sorted_zip)
```

```python
plt.scatter(X, y, color = 'blue', label = 'Actual')
plt.plot(X_sorted, y_sorted, color = 'red', label = 'Polynomial Fit')
plt.xlabel('Mileage(km)')
plt.ylabel('Price')
plt.title('Polynomial Regression(degree 2)');
```

```python
r2_score(y, y_pred)
```

```python
mean_squared_error(y, y_pred)
```
